```python
mca1 =mca.drop(["Unnamed: 0"], axis=1)
mca1
```

```python
mca1.columns = list(map(lambda x: str.lower(x), mca1.columns))

mca1.columns
```

```python
mca1["state"].value_counts(dropna=False)
```

```python
# in this instance we decided to fill NaN with unknown, as 631 values are a 1/6 of Califorina for instance and that change could impare our analysis
```

```python
mca1["response"].value_counts(dropna=False)
```

```python
mca1["response"] = mca1["response"].fillna("No")
mca1["response"].value_counts(dropna=False)
```

```python
mca1["months since last claim"].value_counts(dropna=False)
```

```python
mca1["number of open complaints"].value_counts(dropna=False)
```

```python
# we are going to assign the Nan values to 0.0 as it is the highest number in that column
mca1["number of open complaints"] = mca1["number of open complaints"].fillna("0")
mca1["number of open complaints"].value_counts(dropna=False)
```

```python
mca1["vehicle class"].value_counts(dropna=False)
```

```python
# we are going to assign the Nan values to four-Door class as it is the highest number in that column
mca1["vehicle class"] = mca1["vehicle class"].fillna("Four-Door Car")
mca1["vehicle class"].value_counts(dropna=False)
```

```python
mca1["vehicle size"].value_counts(dropna=False)
```

```python
# we are going to assign the Nan values to Medsize class as it is the highest number in that column
mca1["vehicle size"] = mca1["vehicle size"].fillna("Medsize")
mca1["vehicle size"].value_counts(dropna=False)
```

```python
mca1["vehicle type"].value_counts(dropna=False)
```

```python
# we are going to assign the Nan values to A  as it is the highest number in that column
mca1["vehicle type"] = mca1["vehicle type"].fillna("A")
mca1["vehicle type"].value_counts(dropna=False)
```
